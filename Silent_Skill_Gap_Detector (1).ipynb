```python
#Reading Job Description PDF and Extracting Skills
import PyPDF2
import PyPDF2

file_path = "/content/drive/MyDrive/Silent_Skill_Gap_Detector/data/job_descriptions/Data Scientist - JD and Resume Example 1.pdf"

with open(file_path, "rb") as file:
    reader = PyPDF2.PdfReader(file)

    print("Total Pages:", len(reader.pages))

    first_page = reader.pages[0]
    text = first_page.extract_text()

    print(text)
```

```python
#  - Reading Multiple Job Descriptions
#Combining All Job Description Data
import PyPDF2
import os

folder_path = "/content/drive/MyDrive/Silent_Skill_Gap_Detector/data/job_descriptions"

files = os.listdir(folder_path)

print(files)
```

```python
# Combining All Job Description Data
all_text = ""

for file_name in files:
    file_path = os.path.join(folder_path, file_name)

    with open(file_path, "rb") as file:
        reader = PyPDF2.PdfReader(file)

        for page in reader.pages:
            all_text += page.extract_text() + " "
```

```python
print("Total Characters:", len(all_text))
```

```python
# Skill Extraction from Multiple Job Descriptions
skills = [
    "Python",
    "SQL",
    "Machine Learning",
    "Deep Learning",
    "Statistics",
    "TensorFlow",
    "PyTorch",
    "Data Mining",
    "Predictive Modeling",
    "Cloud Computing"
]
```

```python
found_skills = []

for skill in skills:
    if skill.lower() in all_text.lower():
        found_skills.append(skill)

print(found_skills)
```

```python
print("Total Skills Found:", len(found_skills))
```

```python
# Match Score Calculation
total_skills = len(found_skills)

candidate_match = total_skills - len(missing_skills)

match_score = (candidate_match / total_skills) * 100

print("Match Score:", round(match_score, 2), "%")
```
